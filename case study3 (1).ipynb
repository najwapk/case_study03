# ● The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.


step 1:set up null and alternative hypotheses

H0:There is no increase in sales after stepping into digital marketing
    
   
   H1:There is an increase in the sales after stepping into digital marketing



level of significance is taken 5%,ie 0.05

step2:Decide the test
   
   since the sample data is small and population std deviation is unknown we perform t test

```python
from scipy import statsfrom scipy.stats import ttest_rel
t_value,p_value=stats.ttest_rel(sales.Sales_before_digital_add,sales.Sales_After_digital_add)
print(ttest_rel(sales.Sales_before_digital_add,sales.Sales_After_digital_add))
if(abs(t_value)<p_value):
    print("\033[1msince t_value is less than p_value \nWe fail to reject null hypothesis \nTherefore there is no increase in the sales after stepping into digital marketing\033[0m")
else:
    print("\033[1msince t_value is greater than p_value\nWe reject the null hypothesis \nTherefore there is an increase in the sales after stepping into digital marketing\033[0m")
```

# ● The company needs to check whether there is any dependency between the features “Region” and “Manager”.


```python
from scipy.stats import chi2from scipy.stats import chi2_contingency
table=sales[['Sales_before_digital_add','Sales_After_digital_add']]
table
```

H0: Region and Manager are independent

  H1: Region and Manager are dependent

level of significance is taken as 0.05

```python
stat, p, dof, expected = chi2_contingency(table)
print('dof=%d' % dof)
print(expected)
prob=0.95
critical = chi2.ppf(prob,dof)
print('\033[1mprobability=%.2f\ncritical=%.3f\nstat=%.3f \033[0m' % (prob, critical, stat))
if abs(stat) <= critical:
    print('\033[1msince computed value is less than crictical value \nWe fail to reject the null hypothesis \n  Region and Manager are independent \033[0m')
else:
    print('\033[1msince computed value is greater than crictical value\nWe reject the null hypothesis \nTherefore Region and Manager are dependent \033[0m')
```
